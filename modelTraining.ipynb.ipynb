```python
#Keep only Hyderabad data
df1=df.query("group=='Hyderabad'").reset_index(drop=True)
```

```python
#Check Missing value in data set
df1.isnull().sum(axis=0)
```

```python
#Removing missing value from " ordereditem_product_id " 
df1=df1[pd.notnull(df1['ordereditem_product_id'])]
```

```python
#Converting the string date field to datetime
df1['created']=pd.to_datetime(df1['created'])
```

```python
df1
```

```python
import datetime as dt
# Set the latest date 
Latest_date=dt.datetime(2018,5,5)

```

```python
#Creating RFM Modelling Scores for each customer
rfm_scores=df1.groupby('master_order_id').agg({'created': lambda x:(Latest_date - x.max()).days, 'order_id':lambda x: len(x), 'value':lambda x : x.sum()})


```

```python
#Convert " Created " into type int
rfm_scores['created'] = rfm_scores['created'].astype(int)
```

```python
#Rename column names to recency, Frequency and Monetary
rfm_scores.rename(columns={'created':'Recency',
                          'order_id':'Frequency','value':'Monetary'}, inplace=True)
```

```python
rfm_scores=rfm_scores.reset_index()
```

```python
rfm_scores
```

```python
# K MEAN ALGORITHMS
import matplotlib.pylab as plt
from sklearn.cluster import	KMeans

# Normalization function 
def norm_func(i):
    x = (i - i.min())	/ (i.max() - i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(rfm_scores.iloc[:, 0:])
df_norm.describe()   # min=0 and max=1 by doing Normalization
```

```python
###### scree plot or elbow curve ############
TWSS = []
k = list(range(2, 9))  # we do clusters in between 2 to 9 ie, 2,3,4,5,6,7,8.
k
for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_norm)
    TWSS.append(kmeans.inertia_)
                                   # intertia = Sum of Squared distance of Samples to their closest cluster centre.
TWSS

```

```python

# Scree plot / Elbow Curve = visualize to understand net redcution here.
plt.plot(k, TWSS, 'ro-');plt.xlabel("No_of_Clusters");plt.ylabel("total_within_SS")
```

```python

# Selecting k=3 clusters from the above scree plot which is the optimum number of clusters .
model = KMeans(n_clusters = 4)
model.fit(df_norm)

model.labels_ # getting the labels of clusters assigned to each row 
mb = pd.Series(model.labels_)  # converting numpy array into pandas series object (columns)
mb
```

```python
rfm_scores['clust'] = mb # creating a  new column and assigning it to new column 
rfm_scores
```

```python
## Create pickle file
import pickle
pickle_out = open("credit.pkl","wb")
pickle.dump(rfm_scores,pickle_out)
pickle_out.close()
```
